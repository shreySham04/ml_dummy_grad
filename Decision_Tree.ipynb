```python
import pandas as pd
```

```python
df = pd.read_csv('salaries.csv')
df.head()
```

```python
df.tail()
```

```python
inp=df.drop('salary_more_then_100k',axis='columns')
target=df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l_company=LabelEncoder()
l_job=LabelEncoder()
l_degree=LabelEncoder()
```

```python
inp['company_n']=l_company.fit_transform(inp['company'])
inp['job_n']=l_job.fit_transform(inp['job'])
inp['degree_n']=l_degree.fit_transform(inp['degree'])
```

```python
inp
```

```python
inp
```

```python
inp_n=inp.drop(['company','job','degree'],axis='columns')
```

```python
inp_n
```

```python
target
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inp_n,target)
```

```python
model.score(inp_n,target)
```

```python
model.predict([[1,2,1]])
```

# Tit Anic


```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
inpu=df.drop(['PassengerId','Survived','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
target=df['Survived']
```

```python
inpu
```

```python
target

```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Pclass=LabelEncoder()
le_sex=LabelEncoder()
le_age=LabelEncoder()
le_fare=LabelEncoder()
```

```python
print(inpu.columns)
```

```python
inpu['Pclass_n']=le_Pclass.fit_transform(inpu['Pclass'])
inpu['sex_n']=le_sex.fit_transform(inpu['Sex'])
inpu['age_n']=le_age.fit_transform(inpu['Age'])
inpu['fare_n']=le_fare.fit_transform(inpu['Fare'])
```

```python
inpu
```

```python
inpu_n=inpu.drop(['Pclass','Sex','Age','Fare'],axis='columns')
```

```python
inpu_n
```

```python
inp
```

```python
inpu
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inpu_n,target)
```

```python
model.score(inpu_n,target)
```

```python
model.predict([[0,1,2,24]])
```
